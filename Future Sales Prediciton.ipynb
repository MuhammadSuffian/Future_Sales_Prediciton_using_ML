```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import plotly.express as px
import plotly.graph_objects as go
from sklearn.metrics import r2_score
from sklearn.metrics import root_mean_squared_error
```

```python
data=pd.read_csv("advertising.csv")
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
figure=px.scatter(data,x="Sales",y="Radio",size="Radio", trendline="ols")
figure.show()
```

```python
data.head()
```

```python
x=np.array(data[["TV","Radio","Newspaper"]])
y=np.array(data["Sales"])
# print(x)
# print(y)
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=.2,train_size=.8,random_state=42)
```

```python
model=LinearRegression()
model.fit(xtrain,ytrain)
```

```python
prediction=model.predict(xtest)
```

```python
print(model.score(xtest,ytest))
```

```python
root_square=root_mean_squared_error(ytest,prediction)
r2=r2_score(ytest,prediction)
print("R2:", r2)
print("Root Sqaure",root_square)
```
